```python
import matplotlib.pylab as plt
import numpy as np

n=1000

fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111)

x=np.zeros(n)
y=np.zeros(n)
z=np.zeros(n)
x[0]=0.0
y[0]=0.0
z[0]=0.0
for i in xrange(n-1):
    tmp_x = x
    x[i+1] = 1.76-y[i]*y[i]-0.1*z[i]
    y[i+1] = x[i]
    z[i+1] = y[i]
ax.plot(range(n),x,
    '-',color='black')
        
```

```python
plt.hist(x,bins=64)
```

```python
def leftShift(x,n):
    return x[n:]

fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111)

tau=1
for i, j in zip(x,leftShift(x,tau)):
        ax.plot(i,j,
                'o', color='black')
        
```
